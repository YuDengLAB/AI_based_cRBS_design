```python
import numpy as np
import pandas as pd

data=pd.read_csv("./seq/Fluorescence_intensity_N_0_1129.csv")#read csv file
data=pd.read_csv("./seq/Fluorescence_intensity_N_20_1129.csv")
data=pd.read_csv("./seq/Fluorescence_intensity_R_0_1202.csv")
data=pd.read_csv("./seq/Fluorescence_intensity_R_20_12020.csv")
Flu=data.values.tolist()
print(Flu)
```

```python
import operator
from functools import reduce
Flu=reduce(operator.add,Flu)
Flu=["%.2f" % x for x in Flu]
Flu=np.array(Flu)
print(Flu)
```

```python
np.save('./seq/Fluorescence_intensity_N_0_1129',Flu,allow_pickle=True,fix_imports=True)
np.save('./seq/Fluorescence_intensity_N_20_1129',Flu,allow_pickle=True,fix_imports=True)
np.save('./seq/Fluorescence_intensity_R_0_1202',Flu,allow_pickle=True,fix_imports=True)
np.save('./seq/Fluorescence_intensity_R_20_1202',Flu,allow_pickle=True,fix_imports=True)
```
